```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#READ FILE(Read the CSV file into a DataFrame named df to perform analysis on the GST dataset.)
df = pd.read_csv("C:\\Users\\rabia\\OneDrive\\Desktop\\datasettt.csv")
df
```

```python
#RENAMING COLUMNS(Simplify column names for easier access during analysis )
df.rename(columns={
    'Central Goods and Services Tax ( CGST ) Revenue': 'CGST',
    'State Goods and Services Tax ( SGST )Revenue': 'SGST',
    'Integrated Goods and Services Tax ( IGST )Revenue': 'IGST',
    'CESS Tax Revenue': 'CESS',
    'srcStateName': 'State',
    'Month': 'MonthFull'
}, inplace=True)
df
```

```python
#SNS STYLE
sns.set(style="whitegrid")
```

```python
#BASIC INFO LIKE NULL VALUE,SHAPE,DTYPE
print("Dataset Shape:", df.shape)
print("Column Names:", df.columns.tolist())
print(" Data Types:", df.dtypes)
print(" Missing Values:", df.isnull().sum())
```

```python
#SUMMARY OF STATISTICS(Show statistical summaries for each tax type:Mean, min, max, standard deviation, etc.Helps understand the range and distribution of each tax revenue type.)
print("Descriptive Statistics:")
print(df[['CGST', 'SGST', 'IGST', 'CESS']].describe())
```

```python
# Correlation Heatmap(Visualize the correlation between different types of GST taxes.)
plt.figure(figsize=(8, 6))
corr = df[['CGST', 'SGST', 'IGST', 'CESS']].corr()
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Correlation Heatmap of GST Revenues")
plt.tight_layout()
plt.show()
```

```python
#Revenue Distributions(Understand the frequency and spread of revenue values for each tax type.KDE (Kernel Density Estimate) helps see the smooth distribution curve.)
plt.figure(figsize=(14, 8))
for i, col in enumerate(['CGST', 'SGST', 'IGST', 'CESS']):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[col], bins=30, kde=True, color='skyblue')
    plt.title("{col} Distribution")
plt.tight_layout()
plt.show()
```

```python
#Revenue by State(Identify which states contribute the most to total GST revenue.This helps in economic planning and understanding regional tax strength)
state_totals = df.groupby('State')[['CGST', 'SGST', 'IGST', 'CESS']].sum()
state_totals['Total'] = state_totals.sum(axis=1)
top_states = state_totals.sort_values('Total', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_states['Total'], y=top_states.index)
plt.title("Top 10 States by Total GST Revenue")
plt.xlabel("Total GST Revenue")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
#Monthly Trends(Visualize how GST revenue changes month by month, which can help spot seasonal trends, policy effects, or anomalies.)
monthly_trend = df.groupby('MonthFull')[['CGST', 'SGST', 'IGST', 'CESS']].sum()
monthly_trend = monthly_trend.reset_index()

plt.figure(figsize=(24, 6))
for tax in ['CGST', 'SGST', 'IGST', 'CESS']: 
    sns.lineplot(x='MonthFull', y=tax, data=monthly_trend, label=tax)
plt.title("Monthly GST Revenue Trends")
plt.xticks(rotation=45)
plt.ylabel("Revenue")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Overall GST Type Share(Display the proportional contribution of each GST type to the overall tax revenue collected.)
total_share = df[['CGST', 'SGST', 'IGST', 'CESS']].sum()
plt.figure(figsize=(6, 6))
plt.pie(total_share, labels=total_share.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Overall GST Revenue Share by Type")
plt.tight_layout()
plt.show()
```
